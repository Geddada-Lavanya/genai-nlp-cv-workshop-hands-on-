```python
# Text Preprocessing
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw -1.4')
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
text = "The cats were running faster than the dogs."
tokens = word_tokenize(text)
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]
print(f"Original Text: {text}")
print(f" Lemmatized Words: {lemmatized_words}")
```

```python
#Text Preprocessing
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
text = "Hello Students! This is a simple NLP mini project"
text = text.lower()
text = text.translate(str.maketrans('','',string.punctuation))
tokens = word_tokenize(text)
filtered = [word for word in tokens if word not in stopwords.words('english')]
print("Tokens:",tokens)
print("filtered Words:",filtered)
```

```python
#Word Frequency Counter
from collections import Counter
from nltk.tokenize import word_tokenize
text="NLP is fun. NLP is powerful. Learning NLP is useful."
tokens=word_tokenize(text.lower())
word_count=Counter(tokens)
print(word_count)
print("Most common:",word_count.most_common(3))

```

```python
#Sentiment Analysis
from textblob import TextBlob
text=input("Enter a sentence: ")
analysis=TextBlob(text)
if analysis.sentiment.polarity>0:
  print("positive")
elif analysis.sentiment.polarity<0:
  print("Negative")
else:
  print("Neutral")
```

```python
#Named Entity Recognition
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded SpaceX in the United States"
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

```python
#Spam Email Classifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
#sample data
texts=["free money now", "Hi how are you", "win cash prize", "Let's meet tomorrow"]
labels=[1,0,1,0] # 1=spam, 0=Not Spam

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(texts)

model=MultinomialNB()
model.fit(X,labels)

test=["once try this course"]
test_vec=vectorizer.transform(test)

print("Spam Detection:",model.predict(test_vec))
```

```python
#Text Similarity checker

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

texts=["i love machine Learning", "I like AI and Machine Learning"]
vectorizer=TfidfVectorizer()
tfidf=vectorizer.fit_transform(texts)
similarity=cosine_similarity(tfidf[0:1],tfidf[1:2])
print("Similarity Score: ",similarity[0][0])
```

```python
# Minimum Edit Distance
import nltk
from nltk.metrics import edit_distance

word1="kitten"
word2="sitting"

distance=edit_distance(word1,word2)
print("word1:",word1)
print("word2:",word2)
print("Edit Distance:",distance)
```

```python
#N-Gram(Bigram)
from nltk.util import ngrams
sentence="I love NLP"
tokens=sentence.split()
bigrams=list(ngrams(tokens,2))
print(bigrams)
```

```python
# Stemming (Potter Stemming)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
words=["playing","played","plays"]
for w in words:
  print(ps.stem(w))
```

```python
#maximum entropy Classifier
import nltk
from nltk.classify import MaxentClassifier
train_data=[
    ({"word":"good"},"Positive"),
    ({"word":"excellent"},"Positive"),
    ({"word":"bad"},"Negative"),
    ({"word":"terrible"},"Negative")
]
classifier=MaxentClassifier.train(train_data,max_iter=10)
test_data={"word":"good"}
print("Test Data:",test_data)
print(classifier.classify({"word":"good"}))
print(classifier.classify({"word":"bad"}))
```

```python
 #Hidden Markov Model (sequence prediction)
import nltk
from nltk.tag import hmm

# Training data (list of sentences, each sentence is a list of (word, tag))
train_data = [
    [("I", "PRON"), ("eat", "VERB"), ("mango", "NOUN")],
    [("I", "PRON"), ("eat", "VERB"), ("banana", "NOUN")]
]

# Train the model
trainer = hmm.HiddenMarkovModelTrainer()
model = trainer.train(train_data)

# Test the model
test_sentence = ["I", "eat", "watermelon"]
tagged_output = model.tag(test_sentence)

print(tagged_output)
```
