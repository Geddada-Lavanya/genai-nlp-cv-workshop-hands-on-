```python
import random

human_texts=["I had a great day at college and enjoyed learning new things",
"cyber security is becoming very important in today's world"]
ai_texts=["The rapid evolution of artificial intelligence is transforming industries globally."
"Generative ai models can produce human like text with remarkable experience."]

all_texts=human_texts+ai_texts
random.shuffle(all_texts)

for text in all_texts:
    print("\nText:",text)
    guess=input("Is this ai or Human? ").lower()
    if text in ai_texts and guess =="ai":
        print("Correct!")
    elif text in human_texts and guess =="human":
        print("Correct!")
    else:
        print("OOps! Wrong")
```

```python
#automatic email generation using ai
def generate_email(name,purpose):
  return f"""
Dear {name},
I hope you are doing well. I am writing this email regarding {purpose}
Thank you,
Student
"""
name=input("Enter Recipient name:")
purpose=input("enter Purpose: ")
print(generate_email(name,purpose))

```

```python
#basic Ai chatbot Program
def chatbot(user_input):
  responses={
      "hello": "Hi there! ",
      "what is ai":"AI is intelligence shown by machines",
      "bye":"Goodbye! have a nice day"
  }
  return responses.get(user_input.lower(), "I don't understand")
while True:
  user_input=input("you: ")
  if user_input.lower() == "exit":
    break
  print("Bot:",chatbot(user_input))
```

```python
import random

def generate_story(prompt):
    story_templates = [
        # Emotional Story
        f"Long ago, {prompt}. What seemed ordinary slowly turned into something magical. In the end, it changed everything forever.",

        # Adventure Style
        f"It all began with {prompt}. The journey was dangerous, filled with twists and turns, but the destination was worth every step.",

        # Mystery Style
        f"No one could explain why {prompt}. Strange clues started appearing, leading to a shocking truth nobody expected.",

        # Sci-Fi Style
        f"In the year 3025, {prompt}. Technology had advanced, but this event broke all known rules of science.",

        # Inspirational Story
        f"{prompt} was just the beginning. With courage and determination, a small idea turned into something extraordinary.",

        # Twist Ending
        f"{prompt} seemed normal at first. But in the end, it revealed a secret that changed reality itself."
    ]

    return random.choice(story_templates)

# Main program
prompt = input("Enter a story idea (e.g., 'a robot learning emotions'): ")
story = generate_story(prompt)

print("\nGenerated Story:\n")
print(story)
```
